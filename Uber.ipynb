```python
def distance_calc(longitude1, latitude1, longitude2, latitude2):
    #distance formula
    travel = []
    for pos in range(len(longitude1)):
        long1,lati1,long2,lati2 = map(radians,[longitude1[pos],latitude1[pos],longitude2[pos],latitude2[pos]])
        dist_long = long2 - long1
        dist_lati = lati2 - lati1
        a = sin(dist_lati/2)**2 + cos(lati1) * cos(lati2) * sin(dist_long/2)**2
        c = 2 * asin(sqrt(a))*6371
        travel.append(c)
    return travel
```

```python
df['distance'] = distance_calc(df['pickup_longitude'].to_numpy(),
                                                df['pickup_latitude'].to_numpy(),
                                                df['dropoff_longitude'].to_numpy(),
                                                df['dropoff_latitude'].to_numpy()
                                              )
df.head()
```

```python
def identify_outliers(df1 , col):
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1
    upper_limit = Q3+1.5*IQR
    lower_limit = Q1-1.5*IQR
    df[col] = np.clip(df1[col] , lower_limit , upper_limit)
    return df1

def outliers_treatment(df1 , col_list):
    for c in col_list:
        df1 = identify_outliers(df , c)
    return df1
```

```python
df = outliers_treatment(df , df.iloc[:])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
X=df.drop(['fare_amount'],axis=1)
y=df['fare_amount']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
predictions = lr.predict(X_test)
```

```python
sn.set(rc = {'figure.figsize':(8,8)})
plt.scatter(y_test, predictions)
```

```python
plt.figure(figsize=(8,8))
sn.distplot((y_test-predictions))
```

```python
ax = sn.distplot(y_test, hist=False, color="r", label="Actual Value")
sn.distplot(predictions, hist=False, color="b", label="Fitted Values" , ax=ax)
plt.title('Actual vs Fitted Values for Price')
plt.show()
```

```python
from sklearn import metrics
```

```python
print('MAE:',metrics.mean_absolute_error(y_test,predictions))
print ("RMSE: ", np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
clf= RandomForestRegressor(n_estimators=150)
clf.fit(X_train,y_train)
pred= clf.predict(X_test)
```

```python
ax = sn.distplot(y_test, hist=False, color="r", label="Actual Value")
sn.distplot(pred, hist=False, color="b", label="Fitted Values" , ax=ax)
plt.title('Actual vs Fitted Values for Price')
plt.show()
```
